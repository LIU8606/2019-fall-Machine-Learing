```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
def Univariate_gaussian_data_generator(mean, var, size):

    U = np.random.uniform(0, 1, size)
    V = np.random.uniform(0, 1, size)

    z = (np.sqrt(-2.0 * np.log(U))) * (np.cos(2 * np.pi * V))
    X = z * np.sqrt(var) + mean
    
    return X
```

```python
X = Univariate_gaussian_data_generator(3, 5, 10000)
print(X[0])
#plt.hist(X, bins=100)
#plt.show()
```

```python
def Polynomial_basis_linear_model_data_generator(n, a, w, size):

    e = Univariate_gaussian_data_generator(0,a,size)
    x = np.random.uniform(-1., 1., size)

    A = np.array([[x[index]**i for i in range(n)] for index in range(size)])
    A = A.T
    
    y = np.dot(w,A) + e

    return x, y
```

```python
x, y = Polynomial_basis_linear_model_data_generator(4, 1, [1,2,3,4], 1000)
print(x[0])
#plt.figure(figsize=(20, 10))
#plt.scatter(x, y, alpha=0.5)
#plt.show()
```

```python
Sum = 0
Sumsq = 0
index = 0
last_mean = 0
last_var = 0

X = Univariate_gaussian_data_generator(3, 5, 10000)

while True:
    Sum += X[index]
    Sumsq += X[index]*X[index]
    
    mean = Sum/(index+1)
    
    if index==0:
        var = 0
    else:
        var = (Sumsq - (Sum * Sum) /(index+1)) / index
    
    print("Add data point:", X[index])
    print("Mean = {} Variance = {}".format(mean,var))
    
    if abs(last_mean - mean) < 0.001 and abs(last_var - var) < 0.001:
        break
    if index == 10000:
        print("Can't converage")
        break
    
    last_mean = mean
    last_var = var
    index +=1 
    
```

```python
def Sequential_Estimator(X, index, Sum, Sumsq, last_mean, last_var):
    
    new_Sum = Sum + X[index]
    new_Sumsq = Sumsq + X[index]*X[index]
    
    mean = new_Sum/(index+1)
    
    if index==0:
        var = 0
    else:
        var = (new_Sumsq - (new_Sum * new_Sum) /(index+1)) / index
    
    print("Add data point:", X[index])
    print("Mean = {} Variance = {}".format(mean,var))
    
    if abs(mean - last_mean) < 0.001 and abs(var - last_var) < 0.001:
        return
    if index == 10000:
        print("Can't converage")
        return
    Sequential_Estimator(X, index+1, new_Sum, new_Sumsq, mean, var)
    
```

```python
def Baysian_Linear_regression(x,y,n,b):
    
    last_precision = b*np.identity(n)
    last_mean = np.zeros((n,1))
    index =0
    final = [] # 0 : weight, 1 : var
    ten = [] # 0 : weight, 1 : var
    fifty = [] # 0 : weight, 1 : var

    while True:
        X = x[index]
        Y = y[index]
        
        print("Add data point ({: .5f}, {: .5f})".format(X,Y))
        A = np.array([X**i for i in range(n)]).reshape(1,n)
        target = np.array([Y]).reshape(1, 1)

        precision = 1/a* A.T @ A + last_precision
        
        mean = np.linalg.inv(precision) @ (1/a * A.T @ target + last_precision @ last_mean)

        print("Postirior mean:")
        for i in range(n):
            print('{: .10f}'.format(mean[i][0]))
        
        print()
        print("Posterior variance:")
        variance = np.linalg.inv(precision)
        for i in range(n):
            for j in range(n):
                print("{: .10f}".format(variance[i][j]),end=', ')
            print()
        print()
            
        predictive_distribution_means = (last_mean.T @ A.T).reshape(1)
        predictive_distribution_var = (a + A @ np.linalg.inv(last_precision) @ A.T).reshape(1)

        print("Predictive distribution ~ N({: .5f}, {: .5f})".format(predictive_distribution_means[0],predictive_distribution_var[0]))
        print()
        
        if index == 9:
            ten.append(mean)
            ten.append(precision)
        elif index == 49:
            fifty.append(mean)
            fifty.append(precision)
        elif index == 9999:
            print("Can't converge")
            return [], [], [], 0
            
        if (np.absolute(mean - last_mean) < 1e-6).all():
            final.append(mean)
            final.append(precision)
            return ten, fifty, final, index
        
        last_mean = mean
        last_precision = precision
        index +=1    
```

```python
def draw(plot_pos, draw_x, A, weight, precision, index, title):
    weight = weight.reshape(n)
    draw_y = weight @ A

    plt.subplot(plot_pos)
    plt.ylim(-20,20)
    plt.plot(x[:index], y[:index], '.')
    plt.plot(draw_x, draw_y , 'k')

    var = []

    for i in range(A.shape[1]):
        var.append(a + A[:,i].reshape(1,n) @ np.linalg.inv(precision) @ A[:,i].reshape(n,1))

    var = np.array(var).reshape(A.shape[1])

    plt.plot(draw_x, draw_y + var, 'r')
    plt.plot(draw_x, draw_y - var, 'r')
    plt.title(title)
    
```

```python
b = 1
n = 4
a = 1
w = [1,2,3,4]

x, y = Polynomial_basis_linear_model_data_generator(n, a, w, 10000)
ten, fifty, final, index = Baysian_Linear_regression(x,y,n,b)
```

```python
draw_x = np.linspace(-2.0, 2.0, num=100)
A = np.array([[point**i for i in range(n)] for point in draw_x]).T
draw_y = w @ A

plt.figure(figsize=(15, 10))
plt.subplot(221)
plt.ylim(-20,20)
plt.plot(draw_x, draw_y, 'k')
plt.plot(draw_x, draw_y+a, 'r')
plt.plot(draw_x, draw_y-a, 'r')
plt.title("Groud truth")

draw(222, draw_x, A, final[0], final[1], index, "Predict result")
draw(223, draw_x, A, ten[0], ten[1], 10, "After 10 incomes")
draw(224, draw_x, A, fifty[0], fifty[1], 50, "After 50 incomes")
```
